```python
#Adding the imports
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

#load the datasets
df = pd.read_csv('customer_clustering_data.csv')
df.head()

#we need to extract the features
X = [['Annual Income (k$)', 'Spending Score (1-100)' ]]

#we need to scale the values using the min max scaler 
X_scaled = StandardScaler().fit_transform(X)

#we need to instainiate the model
clustering_model = DBSCAN(eps=0.5 , min_samples=5 )

y_predicted =clustering_model.fit_predict(X_scaled)

#we need to plot the graph
plt.figure(figsize=(30,20))
plt.scatter(c=y_predicted , cmap="virdis")
plt.title("Clustering using the DB-SCAN")
plt.xlabel("Salaries")
plt.ylabel("Customers Spending")
plt.grid()
plt.show()
```
